```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read Excel and show sheet

```python
df = pd.read_excel("used_price.xlsx")
```

```python
df.head()
```

# Price plot

```python
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]+300, y[i], ha = 'center')
```

```python
fig = plt.figure(dpi=800)
ax = fig.add_axes([0,0,1,1])
ax.bar(df['Computer'],df["Price"],color=['red','orangered','silver','dodgerblue','green','blue'])
plt.xlabel('Computer series')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Price (DKK)')
plt.title('Compare different price of my used computers')
# Get average price line
plt.axhline(y=np.mean(df['Price']), color='red', linestyle='--',linewidth=1)
plt.legend(["Average price", "Computer price"])
addlabels(df['Computer'],df["Price"])
plt.show()
```
